# Problem 1 — No Industry Benchmarks

Your revenue leak detection assumes fixed numbers:

```
conversion benchmark = 15%
ROAS benchmark = 2x
retention = 40%
```

But benchmarks **vary heavily by industry**.

Example:

| Industry     | Conversion |
| ------------ | ---------- |
| SaaS         | 5–10%      |
| Ecommerce    | 2–4%       |
| B2B services | 10–20%     |

Without industry context your leaks may be **wrong**.

### Fix

Create:

```
industry_benchmarks.json
```

Example:

```
{
 "saas":{
  "conversion_rate":0.07,
  "roas":3,
  "retention":0.8
 },
 "ecommerce":{
  "conversion_rate":0.03,
  "roas":2.5,
  "retention":0.35
 }
}
```

Then load benchmarks dynamically.
